```python
#Mean Square errror
```

```python
def mse(original,predicted):
    value = np.mean((original-predicted)**2)
    return value
```

```python
x=np.random.randint(1,25,10)
y=np.random.randint(1,25,10)
```

```python
x
```

```python
y
```

```python
error = mse(x,y)
print(error)
```
